## Line Chart Example

```python
import matplotlib.pyplot as plt

def line_chart_example(x, y1,y2=[]):
    plt.figure()
    plt.plot(x, y1,label="$y = x^2$")
    if(y2 != []):
        plt.plot(x, y2,label="$y = x^3$")
    plt.title("Our First Line Chart")
    plt.xlabel("X units")
    plt.ylabel("Y units")
    plt.legend()
    plt.grid(True)
    plt.show()


x = list(range(5))
y1 = [val ** 2 for val in x]

# task: add another to the figure for y = x^3
y2 = [val ** 3 for val in x]
line_chart_example(x,y1,y2)
```
